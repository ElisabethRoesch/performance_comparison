```julia
using DelimitedFiles, Plots
julia_times = readdlm("data/julia_times.txt")[:,1]
r_times = readdlm("data/r_times.txt")[:,1]
python_times =  readdlm("data/python_times.txt")[:,1]
matlab_times = readdlm("data/MATLAB_times.txt")[:,1]
```

```julia
ts = [julia_times, r_times, python_times, matlab_times]
```

```julia
plot(["Test 1.1","Test 1.2","Test 2.1","Test 3.1"], ts[1], label = "Julia", grid = "off", ylab = "Time [sec]")
plot!(["Test 1.1","Test 1.2","Test 2.1","Test 3.1"], ts[2], label = "R")
plot!(["Test 1.1","Test 1.2","Test 2.1","Test 3.1"], ts[3], label = "Python")
plot!(["Test 1.1","Test 1.2","Test 2.1","Test 3.1"], ts[4], label = "MATLAB")
```
